# Titanic Data Cleaning

This notebook was used to handle missing values and create a few simple features.

```python
import pandas as pd

df = pd.read_csv('../data/raw/Titanic-Dataset.csv')
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode().iloc[0])
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
df['FarePerPerson'] = df['Fare'] / df['FamilySize']
df[['FamilySize', 'IsAlone', 'FarePerPerson']].head()
```

```python
df['Sex'] = df['Sex'].str.lower().map({'male': 1, 'female': 0})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
df.head()
```
