# Titanic EDA

This notebook contains the main plots I used before moving the final charts into the app.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/Titanic-Dataset.csv')
```

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival by Class')
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Survived', kde=True, multiple='stack')
plt.title('Age Distribution by Survival')
plt.show()
```

```python
sns.histplot(data=df, x='Fare', hue='Survived', kde=True, multiple='stack')
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
corr = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```
